```python
## Parameter for standard best fit
iniZ_st=0.0126
imf_type_st='chabrier'
imf_bdys_st=[0.1,100] #based on paper p. 18 (and tests on p.19)
imf_yields_range_st=[0.8,100]  #0.8 based on calculations in paper p.19, paper uses 40Msun but then messses up plots, 100Msun good agreement.
sn1a_rate_st='exp' #based on paper .p. 23
exp_dtd_st=2e9  #based on paper p. 23
direct_norm_1a_st=0.016 #better fit #0.02, #based on paper p. 23 (a parameter, see also p22)
dt_st=7e6
transitionmass_st=8 #6.8 #chosen to match the AGB and massive star  intersection.
yield_interp_st='wiersma' # means we correct for initial abundance as in wiersma (independent if yields are net yields or not.)
netyields_on_st=True #tells the code we actual use tabula.ted net yields. 
wiersmamod_st=True
table_st='yield_tables/agb_and_massive_stars_portinari98_marigo01_net_yields.txt'
exclude_masses_st=[] #needs to be empty else 60Msun exlcuded!, cannot justify NOT to use 6,7Msun
sn1a_table_st='yield_tables/sn1a_t03.txt'
iniabu_table_st='yield_tables/iniabu/iniab_solar_Wiersma_test.ppn'
hardsetZ = 0.02
```

## Fig. 3

```python
#%matplotlib nbagg
fign = 1
f, axarr = plt.subplots(3, 2,num=fign)
markevery=10
props = dict(boxstyle='square', facecolor='w', alpha=1)
for specie in ['Total','C','N','O','Si','Fe']:
#specie='Ba-132'
        if specie=='Total':
            plt.sca(axarr[0, 0])
        if specie=='C':
            plt.sca(axarr[0, 1])
        elif specie=='N':
            plt.sca(axarr[1, 0])
        elif specie=='O':
            plt.sca(axarr[1, 1])
        elif specie=='Si':
            plt.sca(axarr[2, 0])
        elif specie=='Fe':
            plt.sca(axarr[2, 1])
            
        if specie == 'Total':
            
            markevery=8            
            s0n.plot_totmasses(fig=fign,source='agb',markevery=markevery,color='b',marker='^',shape=':')
            s0w.plot_totmasses(fig=fign,source='agb',markevery=markevery+5,color='r',marker='>',shape=':')
            s0n.plot_totmasses(fig=fign,source='massive',markevery=markevery,color='b',marker='D')
            s0w.plot_totmasses(fig=fign,source='massive',markevery=markevery+5,color='r',marker='o')
            plt.plot([],[],marker='',linestyle=':',color='k',label='AGBt',linewidth=3)
            plt.plot([],[],marker='',linestyle='-',color='k',label='Massive',linewidth=3)            
            ax=plt.gca()
            l2=ax.lines[-2:]
            plt.legend(l2,('AGB','Massive'),loc=6,prop={'size': 12}).set_visible(True)             
            
        else:  
            s0n.plot_mass(fig=fign,specie=specie,source='agb',norm='no',color='b',label=specie+' (AGB), Z=2e-2',shape=':',marker='^',markevery=markevery)
            s0w.plot_mass(fig=fign,specie=specie,source='agb',norm='no',color='r',label=specie+' (AGB), Z=2e-2',shape=':',marker='>',markevery=markevery+5)
            s0n.plot_mass(fig=fign,specie=specie,source='massive',norm='no',color='b',label=specie+' (massive), Z=2e-2',shape='-',marker='D',markevery=markevery)
            s0w.plot_mass(fig=fign,specie=specie,source='massive',norm='no',color='r',label=specie+' (massive), Z=2e-2',shape='-',marker='o',markevery=markevery+5)   
            plt.legend(loc=6,fontsize=12).set_visible(False)
        if specie in ['C',]:
            plt.ylim(1e-6,1e-2)
        elif specie in ['Si']:
            plt.ylim(1e-5,2e-3)
        elif specie in ['N']:
            plt.ylim(1e-5,2e-3)
        elif specie in ['O']:
            plt.ylim(1e-4,2e-2)       
        elif specie in ['Fe']:
            plt.ylim(1e-5,1e-3)
        ax=plt.gca()
        if specie=='Total':
            plt.plot([],[],marker='^',linestyle=':',color='b',label='AGB')
            plt.plot([],[],marker='>',linestyle=':',color='r',label='Massive')
            #plt.plot([],[],marker='+',linestyle=':',color='r',label='SNIa')
            plt.plot([],[],marker='D',linestyle='-',color='b',label='AGB')
            plt.plot([],[],marker='o',linestyle='-',color='r',label='Massive')
            l2=ax.lines[-4:]
            leg=plt.legend(l2,('AGB, NuGrid','AGB, P98+M01','Massive, NuGrid','Massive, P98+M01'),loc=4,prop={'size': 14})
        ax=plt.gca()    
        if (specie=='N'):
            ax.set_ylabel(r'Accumulated ejecta [M$_{\odot}$]',size=16)
        else:
            plt.ylabel('')
        if specie in ['Total','C','N','O']:
            ax.get_xaxis().set_ticklabels([])
            plt.xlabel('')    
        else:
            plt.xlabel('SSP age [yr]',size=16)
        plt.yscale('log')    
        ax=plt.gca()
        ax.text(0.1, 0.85, specie, transform=ax.transAxes, fontsize=16,verticalalignment='top', bbox=props)  
        if not specie =='Total':
            plt.ylim(1e-5,2e-2)
        else:
            plt.ylim(1e-4,2e0)
#general settings
fig=plt.gcf()        
plt.subplots_adjust(hspace=0.1, bottom=0.125,wspace=0.15,left=0.1,right=0.9)
fig=plt.gcf()        
fig.set_size_inches(14,10,forward=True) #figsize=(6, 4)      
leg.get_frame().set_edgecolor('k')
matplotlib.rcParams.update({'font.size': 16})
```
